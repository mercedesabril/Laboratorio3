### Defino los errores de voltaje y tiempo

```python
def error(vector):
    
    Dms=[]
    for i in range(len(vector)):
        A = str(vector[i])
        ceros = len(A) - A.index('.')-1
        Dms.append(10**(-1*ceros))
    
    return Dms 
```

### Defino la clase Medición, que me deja tener todas las variables de una medición contenidas en un sólo objeto

```python
class Medicion(object):
    
    """
    La clase Medicion es util para cargar las distintas variables de cada una de las mediciones realizadas, con sus valores y unidades. 
    
    Como los archivos no se cargan en orden, self.identificador me permite saber a que medicion corresponde la variable
    """
    
    def __init__(self, **kwargs):
      
      self.df = np.NaN
      self.params = np.NaN
      self.var = np.NaN
      self.error_volt = np.NaN
      self.identificador = np.NaN
      self.corriente = np.NaN
      
      
    def cargar_dataframe(self, dataframe):
      
      self.df = dataframe
      
    def cargar_parametros(self, par, var):
      
      self.params = par
      self.var = var
      
    def cargar_error(self, eV):
    
      self.error_volt = eV  
      
    def cargar_identidad(self, ident):
      
      self.identificador = ident
      
    def cargar_corriente(self, valores):
      
      self.corriente = valores
```

## RC serie

Resistencia = 5000$\Omega$

Frecuencia = 1000Hz

Capacitor = 0.01$\mu \text{F}$

$V_{\text{pico-pico}} = 5.04\text{V}$

```python
# Defino la función de voltaje

def Vcrc (tiempo, v0, vc0, tau):
  
  voltaje_capacitor = vc0 + (v0-vc0)*(1-np.exp(-tiempo/(tau)))
  
  return voltaje_capacitor
```

```python
tau = 5000*0.01*(10**(-6))
tau
```

#### Bajada

```python
df = []
df = pd.read_csv('RC_baj_1000hz.csv', sep = ',', header = None)
df = df.rename(columns = {0:'T', 1:'Vch1', 2:'Vch2'})
med1 = Medicion()
med1.cargar_dataframe(df)
```

```python
med1.cargar_error(error(med1.df['Vch2'].values))
```

```python
params, var = curve_fit(Vcrc, (med1.df['T'].values-0.0005), med1.df['Vch2'].values, p0 = [0, 5,5*10**(-5)], sigma=med1.error_volt)
med1.cargar_parametros(params, var)
print('Los parámetros v0, vc0 y tau encontrados según el ajuste son: ',  med1.params, ' y el error correspondiente:', np.sqrt(np.diag(med1.var)))
```

```python
fig = plt.figure(figsize=(10, 7))

ax = plt.gca()

med1.df.plot(kind='line',x='T',y='Vch1',ax=ax, color = 'black')
med1.df.plot(kind='line',x='T',y='Vch2', color='gray', ax=ax, lw = 6)
plt.plot(np.linspace(0.00,0.001,50000)+0.0005, Vcrc(np.linspace(0.00,0.001,50000), *med1.params), label = 'Ajuste', lw = 3, color = 'black', linestyle = ':')
plt.title('Transitorios en circuito RC. R = 5000ohms, C = 10nF, f = 1000Hz. \n Flanco de bajada')
plt.legend()
```

#### Defino la función corriente para el ajuste

```python
def Isub2 (tiempo, alpha, w0, A, C):
  
  arg = (w0**2) - (alpha**2)
  
  corriente = np.sin(tiempo*np.sqrt(arg)) * np.exp(-tiempo*alpha) * A + C
  
  return corriente
```

```python
plt.plot(np.linspace(0, 0.005, 50000)+0.010, Isub2(np.linspace(0, 0.005, 50000), 1000,10000,7.96*10**(-5), 0.00003))
```

```python
def Isobre (tiempo, alpha, w0):
  
  arg = (alpha**2) - (w0**2)
  
  V0 = 5
  
  offvert = 0.38
  
  corriente = np.sinh(tiempo*np.sqrt(arg)) * np.exp(-tiempo*alpha) * V0/np.sqrt(arg) *offvert
  
  return corriente
```

```python
plt.plot(np.linspace(0.025, 0.05, 50000)-0.0009, Isobre(np.linspace(0,0.025, 50000), 1150, 1000))
```

```python
fig = plt.figure(figsize=(10, 7))

ax = plt.gca()

plt.plot(med11.df['T'].values, med11.corriente, lw = 6, color = "gray", label = "Ich2")
plt.plot(np.linspace(0.025, 0.05, 50000)-0.0009, Isobre(np.linspace(0,0.025, 50000), 1150, 900), label = 'Curva teórica', lw = 3, color = 'black', linestyle = ':')
plt.title('Transitorios en circuito RLC. R = 2300ohms, L = 1H, C = 1000nF, f = 20Hz.')
plt.ylabel('Corriente I(A)')
plt.xlabel('Tiempo (seg)')
plt.legend()
```

```python
def Icrit(tiempo, alpha):
  
  V0 = 5
  
  L = 1
  
  corriente = (V0/L)*tiempo*np.exp(-1*tiempo*alpha)
  
  return(corriente)
  
```

```python
fig = plt.figure(figsize=(10, 7))

ax = plt.gca()

plt.plot(med13.df['T'].values, med13.corriente, lw = 6, color = "gray", label = "Ich2")
plt.plot(np.linspace(0.05, 0.08, 50000)-0.003, -1*Icrit(np.linspace(0,0.025, 50000), 1150)+0.0006, label = 'Curva teórica', lw = 3, color = 'black', linestyle = ':')
plt.title('Transitorios en circuito RLC crítico. R = 2300ohms, L = 1H, C = 1000nF, f = 20Hz.')
plt.ylabel('Corriente I(A)')
plt.xlabel('Tiempo (seg)')
plt.legend()
```
